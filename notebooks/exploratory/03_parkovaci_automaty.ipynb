```python
import geopandas as gpd

df = gpd.read_file("../../data/parkovaci_automaty/parkovaci_automaty.geojson")

df.head()
```

```python
import plotly.graph_objects as go # or plotly.express as px
fig = go.Figure() # or any Plotly Express function e.g. px.bar(...)

df["hover_text"] = (
    "<b>PA:</b> " + df["pa"].astype(str) + "<br>" +
    "<b>PX:</b> " + df["px"].astype(str) + "<br>" +
    "<b>Street:</b> " + df["street"].fillna("") + "<br>" +
    "<b>Code:</b> " + df["code"].astype(str) + "<br>"
)

# Calculate bounds and center
min_lat, max_lat = df["geometry"].y.min(),  df["geometry"].y.max()
min_lon, max_lon = df["geometry"].x.min(), df["geometry"].x.max()

center_lat = (min_lat + max_lat) / 2
center_lon = (min_lon + max_lon) / 2

fig = go.Figure()

trace = go.Scattermap(
    lon=df["geometry"].x,
    lat=df["geometry"].y,
    mode="markers",
    text=df["hover_text"],
    hoverinfo="text",
    marker=dict(size=9, color="royalblue", opacity=0.8),
)

fig.add_trace(trace)

fig.update_layout(
    title="Parkovací automaty",
    map=dict(
        center=dict(lat=center_lat, lon=center_lon),
        zoom=10
    ),
    height=800,
)

fig.show()
```
