```python
import geopandas as gpd

df = gpd.read_file("../../data/objekty_mestske_policie_praha/objekty_mestske_policie_praha.geojson")

df.head()
```

```python
import plotly.graph_objects as go # or plotly.express as px
fig = go.Figure() # or any Plotly Express function e.g. px.bar(...)

df["hover_text"] = (
    "<b>NKU:</b> " + df["nku"].astype(str) + "<br>" +
    "<b>NVPK:</b> " + df["nvpk"].astype(str) + "<br>" +
    "<b>Poznamka:</b> " + df["pozn"].fillna("") + "<br>" +
    "<b>ID poskytovatele:</b> " + df["id_poskyt"].astype(str) + "<br>" +
    "<b>OR MP:</b> " + df["or_mp"].astype(str)
)

# Calculate bounds and center
min_lat, max_lat = df["lat"].min(), df["lat"].max()
min_lon, max_lon = df["lon"].min(), df["lon"].max()

center_lat = (min_lat + max_lat) / 2
center_lon = (min_lon + max_lon) / 2

fig = go.Figure()

trace = go.Scattermap(
    lon=df["lon"],
    lat=df["lat"],
    mode="markers",
    text=df["hover_text"],
    hoverinfo="text",
    marker=dict(size=9, color="royalblue", opacity=0.8),
)

fig.add_trace(trace)

fig.update_layout(
    title="Objekty Městské policie Praha",
    map=dict(
        center=dict(lat=center_lat, lon=center_lon),
        zoom=10
    ),
    height=800,
)

fig.show()
```
