```python
import geopandas as gpd

from utils.geospatial_utils import extract_polygons

df_police = gpd.read_file("../../data/objekty_mestske_policie_praha/objekty_mestske_policie_praha.geojson")
df_districts = gpd.read_file("../../data/mestske_casti/mestske_casti.geojson")
```

```python
from shapely.geometry import shape

df_districts = gpd.read_file("../../data/mestske_casti/mestske_casti.geojson")

districts_geometries = {}
for index, row in df_districts.iterrows():
    geom = shape(row["geometry"]).buffer(0)
    districts_geometries[row["nazev_1"]] = geom
```

```python
# Get number of points inside polygons
for name, geom in districts_geometries.items():
    count = df_police["geometry"].apply(geom.contains).sum()
    print(f"{name}: {count}")
```
